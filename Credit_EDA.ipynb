# Exploratory Data Analysis

Basic Setup and Loading Data

```python
import pandas as pd
import numpy as np

data = pd.read_csv('credit_risk_dataset (2).csv')

data.head()
```

Basic Information

```python
data.info()

data.describe()

data.isnull().sum()

data.dtypes
```

 Distribution and Relationships

```python

import seaborn as sns
import matplotlib.pyplot as plt

```

```python
sns.pairplot(data)
plt.show()

```

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()

```

```python
data.hist(figsize=(12, 10), bins=20)
plt.show()
```

5. Handling Missing Data

```python
# Visualize missing values
sns.heatmap(data.isnull(), cbar=False)
plt.show()

```

```python
# Drop rows with missing values
data_cleaned = data.dropna()

```

```python

# Fill missing values with mean/median
data_filled = data.fillna(data.mean())
```

Outliers

```python
# Boxplot for numeric columns
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

Feature Distribution

```python
# Density plot
for col in numeric_columns:
    sns.kdeplot(data[col], shade=True, label=col)
    plt.legend()
    plt.title(f'Density plot for {col}')
    plt.show()

```

```python
# Count plot for categorical columns
categorical_columns = data.select_dtypes(include=['object']).columns
for col in categorical_columns:
    sns.countplot(y=data[col])
    plt.title(f'Count plot for {col}')
    plt.show()
```
